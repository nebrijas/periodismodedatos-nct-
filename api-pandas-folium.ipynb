Las variables con las que vamos a trabajar en este caso son 3: la **url** de la web de Zaragoza, las **coordenadas** de la provincia y el **mapa** de esta. 

Comenzamos insertando la URL de los accidentes de tráfico de la provincia. 
- Añadimos .csv para adjuntar los datos en una tabla formato excel que visualizaremos posteriormente. 
- Usamos rows=20 para delimitar las filas a 20.

Después añadiremos las coordenadas de Zaragoza y posteriormente usaremos la función folium.Map + la ubicación de las coordenadas de la misma para la creación de nuestro mapa. Con la orden 'mapa' se imprimirá el mapa de Zaragoza a continuación. 


```python
url = 'https://www.zaragoza.es/sede/servicio/transporte/accidentalidad-trafico/accidente.csv?rows=20'
coords_zrgz = [41.649693,-0.887712]
mapa = folium.Map(location=coords_zrgz)
```

A continuación realizaremos un bucle for para el dataframe geometry, es decir, para las longitudes y latitudes del mapa. Separamos los datos de las coordenadas con una coma con la función split y con la += [float(punto_coord[1]) haremos que se vayan añadiendo todos los datos (x +1) de todas las longitudes del mapa. Imprimimos la tabla de datos con la función longitudes.

Par el bucle for de la latitud haremos los mismos pasos pero sustituimos el número 1 por 0. Veremos los datos en forma de lista. 

```python
for i in df['geometry']: 
    punto_coord = i.split(',')
    longitudes += [float(punto_coord[1])]
longitudes
```

```python
for i in df['geometry']: 
    punto_coord = i.split(',')
    latitudes += [float(punto_coord[0])]
latitudes
```

En el siguiente paso crearemos una tabla con los datos de las coordenadas (**lat, long**) de Zaragoza. 
Una vez hayamos impreso esos datos usaremos la función **concatenar** para unir los datos del eje 1 e imprimir el tipo de accidente y sus coordenadas

```python
df_coord = pd.DataFrame({'long':longitudes,'lat':latitudes})
df_coord
```

```python
df_accidentes= pd.concat([df['type'],df_coord],axis=1)
df_accidentes
```

Ahora vamos a crear un marcador en el mapa con las coordenadas de Zaragoza. El marcador lo crearemos a través de la función **folium.Marker** y le añadiremos un pop up que diga "Hola Zaragoza" cuando hagamos click en el marcador. Con la función **add_ child** veremos el icono representado en el mapa por encima del resto de elementos, 



```python
marcador = folium.Marker(coords_zrgz, popup="¡Hola, Zaragoza!")
mapa = mapa.add_child(marcador)
mapa
```

```python
for indice, fila in df_accidentes.iterrows():
    marcador = folium.Marker([fila['lat'],fila['long']],\
                            popup=fila['type'])
    mapa.add_child(marcador)
mapa
```

Como podemos comprobar, con este código **no aparecen los distintos pop ups de los accidentes**.
